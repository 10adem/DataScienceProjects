```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("users.data", sep="\t", names=["user_id", "item_id", "rating", "timestamp"])
```

```python
data.head()
```

```python
len(data)
```

```python
filmler = pd.read_csv("movie_id_titles.csv")
```

```python
filmler.head()
```

```python
len(filmler)
```

```python
data = pd.merge(data, filmler, on = "item_id")
```

```python
data.head()
```

```python
data.tail()
```

```python
tablo = data.pivot_table(index = "user_id", columns= "title", values = "rating")
tablo.head()
```

```python
starwars = tablo["Star Wars (1977)"]
starwars.head()
```

```python
type(starwars)
```

```python
korelasyon = tablo.corrwith(starwars)
```

```python
korelasyon
```

```python
kore = pd.DataFrame(korelasyon, columns=["Correlation"])
kore.dropna(inplace=True)
kore.head()
```

```python
kore.sort_values("Correlation", ascending=False).head(15)
```

```python
data.head()
```

```python
data.drop((["timestamp"]), axis=1)
```

```python
ratings = pd.DataFrame(data.groupby("title")["rating"].mean())
ratings.sort_values("rating", ascending=False).head()
```

```python
ratings["oy_sayisi"] = pd.DataFrame(data.groupby("title")["rating"].count())
```

```python
ratings.head()
```

```python
ratings.sort_values("oy_sayisi", ascending=False).head()
```

```python
kore.head()
```

```python
kore = kore.join(ratings["oy_sayisi"])
kore.head()
```

```python
kore[kore["oy_sayisi"] > 100].sort_values("Correlation", ascending=False).head()
```
