```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Employee.csv")
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(18, 12))

# 1. Eitim Durumuna Gre Ayrlma Oran

plt.subplot(2, 3, 1)
education_plot = sns.countplot(x="Education", hue="LeaveOrNot", data=df, palette="Set2")
plt.title("Eitim Durumuna Gre Ayrlma Dalm")
plt.xlabel("Eitim Seviyesi")
plt.ylabel("alan Says")
for container in education_plot.containers:
    education_plot.bar_label(container, fmt="%.0f")

plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(18, 12))

# 2. ehre Gre Ayrlma Oran

plt.subplot(2, 3, 2)
city_plot = sns.countplot(x="City", hue="LeaveOrNot", data=df, palette="Set2", order=df["City"].value_counts().index)
plt.title("ehre Gre Ayrlma Dalm")
plt.xlabel("ehir")
plt.ylabel("alan Says")
plt.xticks(rotation=45)
for container in city_plot.containers:
    city_plot.bar_label(container, fmt="%.0f")

plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(18, 12))

# 3. Maa Kademesine Gre Ayrlma Oran

plt.subplot(2, 3, 3)
tier_plot = sns.countplot(x="PaymentTier", hue="LeaveOrNot", data=df, palette="Set2")
plt.title("Maa Kademesine Gre Ayrlma Dalm")
plt.xlabel("Maa Kademesi")
plt.ylabel("alan Says")
for container in tier_plot.containers:
    tier_plot.bar_label(container, fmt="%.0f")

plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(18, 12))

# 4. Cinsiyete Gre Ayrlma Oran

plt.subplot(2, 3, 4)
gender_plot = sns.countplot(x="Gender", hue="LeaveOrNot", data=df, palette="Set2")
plt.title("Cinsiyete Gre Ayrlma Dalm")
plt.xlabel("Cinsiyet")
plt.ylabel("alan Says")
for container in gender_plot.containers:
    gender_plot.bar_label(container, fmt="%.0f")

plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(18, 12))

# 5. Daha nce Atamasz Kalma Durumuna Gre Ayrlma Oran

plt.subplot(2, 3, 5)
bench_plot = sns.countplot(x="EverBenched", hue="LeaveOrNot", data=df, palette="Set2")
plt.title("Atamasz Kalma Durumuna Gre Ayrlma Dalm")
plt.xlabel("Daha nce Atamasz Kald m?")
plt.ylabel("alan Says")
for container in bench_plot.containers:
    bench_plot.bar_label(container, fmt="%.0f")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))

# 1. Ya Dalm (Ayrlanlar ve Kalanlar)

plt.subplot(2, 3, 1)
sns.histplot(data=df, x="Age", hue="LeaveOrNot", kde=True, element="step", palette="Set2")
plt.title("Ya Dalm (Ayrlan vs Kalan)")
plt.xlabel("Ya")
plt.ylabel("Frekans")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))

# 2. Mevcut Alandaki Deneyim Dalm

plt.subplot(2, 3, 2)
sns.histplot(data=df, x="ExperienceInCurrentDomain", hue="LeaveOrNot", kde=True, element="step", palette="Set2", discrete=True)
plt.title("Deneyim Dalm (Ayrlan vs Kalan)")
plt.xlabel("Mevcut Alandaki Deneyim (Yl)")
plt.ylabel("Frekans")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))

# 3. e Giri Yl Dalm

plt.subplot(2, 3, 3)
sns.histplot(data=df, x="JoiningYear", hue="LeaveOrNot", kde=True, element="step", palette="Set2", discrete=True)
plt.title("e Giri Yl Dalm (Ayrlan vs Kalan)")
plt.xlabel("e Giri Yl")
plt.ylabel("Frekans")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))

# 4. Saysal Deikenlerin Korelasyon Matrisi

plt.subplot(2, 3, (4,6))
numeric_df = df[["Age", "ExperienceInCurrentDomain", "JoiningYear", "PaymentTier", "LeaveOrNot"]]
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", center=0, square=True)
plt.title("Saysal Deikenlerin Korelasyon Matrisi")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(3, 3, figsize=(18, 15))
fig.suptitle('alan Ayrlma Analizi - zet Grafikler', fontsize=16, y=1.02)

sns.countplot(ax=axes[0, 0], x='Education', hue='LeaveOrNot', data=df)
axes[0, 0].set_title('Eitim Durumu')
axes[0, 0].tick_params(axis='x', rotation=45)

sns.countplot(ax=axes[0, 1], x='City', hue='LeaveOrNot', data=df, order=df['City'].value_counts().index)
axes[0, 1].set_title('ehir')
axes[0, 1].tick_params(axis='x', rotation=45)

sns.countplot(ax=axes[0, 2], x='PaymentTier', hue='LeaveOrNot', data=df)
axes[0, 2].set_title('Maa Kademesi')

sns.countplot(ax=axes[1, 0], x='Gender', hue='LeaveOrNot', data=df)
axes[1, 0].set_title('Cinsiyet')

sns.countplot(ax=axes[1, 1], x='EverBenched', hue='LeaveOrNot', data=df)
axes[1, 1].set_title('Atamasz Kald m?')

sns.histplot(ax=axes[1, 2], data=df, x='Age', hue='LeaveOrNot', kde=True, element='step')
axes[1, 2].set_title('Ya Dalm')

sns.histplot(ax=axes[2, 0], data=df, x='ExperienceInCurrentDomain', hue='LeaveOrNot', kde=True, element='step', discrete=True)
axes[2, 0].set_title('Deneyim Dalm')

sns.histplot(ax=axes[2, 1], data=df, x='JoiningYear', hue='LeaveOrNot', kde=True, element='step', discrete=True)
axes[2, 1].set_title('e Giri Yl')

sns.heatmap(ax=axes[2, 2], data=corr_matrix, annot=True, cmap='coolwarm', center=0, square=True, fmt='.2f')
axes[2, 2].set_title('Korelasyon Matrisi')

plt.tight_layout()
plt.show()
```
